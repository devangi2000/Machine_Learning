```python
#categorical features are converted to dummy variables--like one-hot encoding
df = pd.get_dummies(dataset, columns=['sex', 'cp', 'fbs', 'restecg','exang', 'slope', 'ca', 'thal'])
```

```python
df
```

```python
print(df.shape)
print(dataset.shape)
```

```python
scaler = StandardScaler()
cols_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
df.head()
```

```python
X = df.drop(['target'], axis=1)
y = df['target']
scores = []
for k in range(1,100,10):
    rfc = RandomForestClassifier(n_estimators=k)
    score=cross_val_score(rfc,X,y,cv=10)
    print(score)
    print(score.mean())
    scores.append(score.mean())
```

```python
plt.figure(figsize=(20,20))
plt.plot([k for k in range(1,400,40)], scores, color = 'red')
#for i in range(1,100,10):
 #   plt.text(i, scores[i-1], (i, scores[i-1]))
plt.xticks([i for i in range(1,400,40)])
plt.xlabel('Number of Estimators (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
#now remake the model with k=240
clf = RandomForestClassifier(n_estimators=240)
score = cross_val_score(clf, X,y,cv=10)
print(score.mean())
```
