```python
#categorical features are converted to dummy variables--like one-hot encoding
df = pd.get_dummies(dataset, columns=['sex', 'cp', 'fbs', 'restecg','exang', 'slope', 'ca', 'thal'])
```

```python
print(df.shape)
print(dataset.shape)
df
```

```python
scaler = StandardScaler()
cols_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
```

```python
df.head()
```

```python
X = df.drop(['target'], axis=1)
y = df['target']
scores = []
for k in range(1,100,10):
    dt = DecisionTreeClassifier(criterion='gini',max_depth=k)
    score=cross_val_score(dt,X,y,cv=10)
    print(score)
    print(score.mean())
    scores.append(score.mean())
```

```python
plt.figure(figsize=(20,20))
plt.plot([x for x in range(1,100,10)], scores, color='red')
plt.xlabel('Depth')
plt.ylabel('Scores')
plt.title('Scores plotted against various depth values')
plt.xticks([x for x in range(1,100,10)])
```

```python
clf = DecisionTreeClassifier(criterion='gini',max_depth=91)
score = cross_val_score(clf,X,y,cv=10)
print(score.mean())
```

```python
X = df.drop(['target'], axis=1)
y = df['target']
scores = []
for k in range(1,100,10):
    dt = DecisionTreeClassifier(criterion='entropy',max_depth=k)
    score=cross_val_score(dt,X,y,cv=10)
    print(score)
    print(score.mean())
    scores.append(score.mean())
```

```python
plt.figure(figsize=(20,20))
plt.plot([x for x in range(1,100,10)], scores, color='red')
plt.xlabel('Depth')
plt.ylabel('Scores')
plt.title('Scores plotted against various depth values')
plt.xticks([x for x in range(1,100,10)])
```
